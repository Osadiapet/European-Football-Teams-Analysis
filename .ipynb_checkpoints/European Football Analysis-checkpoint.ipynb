```python
# - We're calculating and creating two new columns to analyze football match outcomes and goal differences.
# - 'HomeTeamGD' represents the goal difference for the home team in each match.
# - 'AwayTeamGD' represents the goal difference for the away team in each match.
# - We also create a 'Draw' column to identify matches that ended in a draw.
# - The 'Draw' column is set to 'True' if the match is a draw (both teams scored the same number of goals).


#creating a new column for home and away teams goals difference
matches_data['HomeTeamGD']=matches_data['home_team_goal']-matches_data['away_team_goal']  
matches_data['AwayTeamGD']=matches_data['away_team_goal']-matches_data['home_team_goal']
#creating a column for drwan matches 
matches_data['Draw']=matches_data['home_team_goal']==matches_data['away_team_goal']
```

```python
# - We are creating additional columns in the 'matches_data' DataFrame to analyze and categorize match results and points for teams.
# - 'HomeResults' and 'AwayResults' columns are added to indicate the match result for the home and away teams ('W' for win, 'D' for draw, 'L' for lose).
# - 'HTPoints' and 'ATPoints' columns calculate the points earned by teams in a match (3 points for a win, 1 point for a draw, 0 points for a loss).
# - 'HomeWin' and 'AwayWin' columns count the number of wins for the home and away teams, where 1 represents a win and 0 represents other results.

#adding a column for teams match status 
def match_status(initial = 0):
    if initial > 0:
        return 'W'
    elif initial == 0:
        return 'D'
    else:
        return 'L'
matches_data['HomeResults'] = matches_data['HomeTeamGD'].apply(match_status)
matches_data['AwayResults'] = matches_data['AwayTeamGD'].apply(match_status)

#calculating points for teams (win=3, draw=1, lose=0)
def match_status2(initial2 = 'w'):
    if initial2 =='W':
        return 3
    elif initial2 =='D':
        return 1
    else:
        return 0
matches_data['HTPoints'] = matches_data['HomeResults'].apply(match_status2)
matches_data['ATPoints'] = matches_data['AwayResults'].apply(match_status2)

#counting the number of wins for each team
def check_wins(wins='W'):
    
    if wins=='W':
        
        return 1
    else:
        return 0
    
matches_data['HomeWin'] = matches_data['HomeResults'].apply(check_wins)
matches_data['AwayWin'] = matches_data['AwayResults'].apply(check_wins)
```

Here, we will perform statistical test to see how these team attributes contributes to the outcome of the match results. 
The aim is is to investigate how different team playing attributes are associated with match results (Win, Draw, Loss). In other words, we want to determine if there is a statistically significant relationship between specific team attributes and the outcomes of football matches. The Chi-Square tests will help us understand whether these attributes have an impact on whether a team wins, draws, or loses a match.

- Teams focusing on "Mixed" passing have a higher home win percentage (27.04%) compared to "Long" (5.53%) or "Short" (13.40%) passing strategies.

These observations provide valuable insights into the potential impact of various team playing attributes on match outcomes. However, to draw conclusive findings, we will subject these relationships to statistical significance testing to determine if these associations are indeed significant or if they could be attributed to chance."

```python
# List of independent variables
independent_vars = ['buildUpPlaySpeedClass', 'buildUpPlayDribblingClass', 'buildUpPlayPassingClass',
                    'buildUpPlayPositioningClass', 'chanceCreationPassingClass', 'chanceCreationCrossingClass',
                    'chanceCreationShootingClass', 'chanceCreationPositioningClass', 'defencePressureClass',
                    'defenceAggressionClass', 'defenceTeamWidthClass', 'defenceDefenderLineClass']

# Create subplots for each crosstabulation
plt.figure(figsize=(20, 16))
plt.suptitle("Impact of Variables on AwayResults", fontsize=24, fontweight='bold')  # Overall title

for i, var in enumerate(independent_vars, 1):
    plt.subplot(3, 4, i)
    crosstab = pd.crosstab(matches_home[var], matches_home['AwayResults'], normalize=True) * 100
    sns.heatmap(crosstab, annot=True, cmap="YlGnBu")
    plt.title(f'{var} vs. AwayResults')
    plt.xlabel('AwayResults')
    plt.ylabel(var)

plt.tight_layout(rect=[0, 0, 1, 0.96])  
plt.show()
```

### Chi-Square Test

We want to investigate how different team playing attributes relate to the HomeResults, which includes three categories (W, L, D). 

**Null Hypothesis (H0):**

There is no significant relationship between team playing attributes (e.g., buildUpPlaySpeedClass, buildUpPlayDribblingClass, etc.) and the HomeResults (W, L, D). In other words, the distribution of HomeResults is independent of team playing attributes.

**Alternative Hypothesis (H1):**

There is a significant relationship between team playing attributes and the HomeResults (W, L, D). The distribution of HomeResults is not independent of team playing attributes.

We wil be testing these hypotheses for each attribute to determine whether they are associated with the outcome of a match (HomeResults). If the p-value for a chi-square test is less than your chosen significance level (0.05), you would reject the null hypothesis for that attribute, indicating a significant relationship between that attribute and HomeResults.

```python

# List of independent variables
independent_vars = [
    'buildUpPlaySpeedClass', 'buildUpPlayDribblingClass', 'buildUpPlayPassingClass',
    'buildUpPlayPositioningClass', 'chanceCreationPassingClass', 'chanceCreationCrossingClass',
    'chanceCreationShootingClass', 'chanceCreationPositioningClass', 'defencePressureClass',
    'defenceAggressionClass', 'defenceTeamWidthClass', 'defenceDefenderLineClass'
]
# Create an empty DataFrame to store the results
results = pd.DataFrame(columns=['Variable', 'Chi-Square Statistic', 'Degrees of Freedom', 'P-Value'])

```

```python
# Create an empty list to store the results
results = []

import scipy.stats as stats

# Loop through independent variables and perform chi-square tests
for var in independent_vars:
    # Create a contingency table
    contingency_table = pd.crosstab(matches_home[var], matches_home['HomeResults'])
    
    # Perform chi-square test
    chi2, p, dof, _ = stats.chi2_contingency(contingency_table)
    
    # Append results to the list
    results.append({
        'Variable': var,
        'Chi-Square Statistic': chi2,
        'Degrees of Freedom': dof,
        'P-Value': p
    })

# Convert the list to a DataFrame
results_df = pd.DataFrame(results)

# Display the results
results_df
```

**Conclusion:**

The Chi-Square tests performed on various team playing attributes have provided insights into their statistical significance in relation to match outcomes (Win, Draw, Loss). The results indicate that, all the following attributes have shown statistically significant associations with match results:

- `buildUpPlayDribblingClass`
- `buildUpPlayPassingClass`
- `buildUpPlayPositioningClass`
- `chanceCreationPassingClass`
- `chanceCreationCrossingClass`
- `chanceCreationShootingClass`
- `chanceCreationPositioningClass`
- `defencePressureClass`
- `defenceAggressionClass`

This suggests that these attributes are not influenced by mere chance, but they have a meaningful impact on whether a team wins, draws, or loses a match.

### Analyzing Football Player Attributes Using Factor Analysis (FA)

Factor analysis is a powerful statistical technique used to explore the underlying structure and relationships among a set of observed variables. FA analysis can provide valuable insights into the latent factors that may influence a player's performance. 

Description:

Football is a game of skill, precision, and teamwork. To gain a deeper understanding of what makes a player truly exceptional, we would look into the  player attributes. From their dribbling finesse to their goalkeeping prowess, footballers possess a wide range of skills and abilities.

Factor Analysis is our key to unlocking the hidden patterns within these attributes. Our goal is to simplify the complexity of player attributes and discover the latent factors that drive performance on the field. By identifying these underlying factors, we aim to gain insights into the essential skills and characteristics that contribute to a player's success.

We will analyze attributes such as crossing, finishing, sprint speed, and goalkeeping reflexes. Through Factor Analysis, we'll uncover whether these attributes can be grouped into common factors, providing a more comprehensive perspective on player capabilities. Are there inherent skills that define an exceptional forward, midfielder, or defender? Our analysis seeks to answer these questions and offer a fresh perspective on the intricate world of football player attributes.

By revealing the underlying factors, we aim to help football enthusiasts, coaches, and analysts better understand what truly matters on the pitch. Factor Analysis can provide a structured approach to evaluating players, informing recruitment strategies, and enhancing overall team performance.

```python
fa_attributes = players_data[['crossing', 'finishing',
       'heading_accuracy', 'short_passing', 'volleys', 'dribbling', 'curve',
       'free_kick_accuracy', 'long_passing', 'ball_control', 'acceleration',
       'sprint_speed', 'agility', 'reactions', 'balance', 'shot_power',
       'jumping', 'stamina', 'strength', 'long_shots', 'aggression',
       'interceptions', 'positioning', 'vision', 'penalties', 'marking',
       'standing_tackle', 'sliding_tackle', 'gk_diving', 'gk_handling',
       'gk_kicking', 'gk_positioning', 'gk_reflexes']]
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit and transform the data
scaled_fa_attributes = scaler.fit_transform(fa_attributes)

# Create a new DataFrame with the scaled data
scaled_fa_attributes_df = pd.DataFrame(scaled_fa_attributes, columns=fa_attributes.columns)

# Display the first few rows of the scaled data
scaled_fa_attributes_df.head()
```

```python
# Instantiate factor analysis object
from factor_analyzer.factor_analyzer import FactorAnalyzer 
fa = FactorAnalyzer(rotation='varimax')
fa.fit(scaled_fa_attributes_df)
# Check Eigenvalues
ev, v = fa.get_eigenvalues()
ev
```

We are employing the Kaiser Criterion for selecting the number of factors in our factor analysis. According to this criterion, we will retain factors for which the eigenvalues are greater than 1. Eigenvalues measure the variance explained by each factor, and an eigenvalue exceeding 1 signifies that the factor accounts for more variance than an individual observed variable. By adopting this criterion, we aim to identify and keep factors that capture a substantial amount of information in our data, aiding our analysis in deriving meaningful insights.

```python
# Select the player attributes
attributes = players_num # Include all your attributes

# Ensure the data is suitable for FA (e.g., no missing values)

# Perform Factor Analysis
fa = FactorAnalyzer(rotation='varimax', n_factors=5)  # Adjust the number of factors as needed
fa.fit(scaled_fa_attributes_df)

# Get factor loadings
factor_loadings = fa.loadings_

# Create a DataFrame to display factor loadings
factor_loadings_df = pd.DataFrame(factor_loadings, index=scaled_fa_attributes_df.columns, columns=['Factor 1', 'Factor 2', 'Factor 3', 'Factor 4', 'Factor 5'])
factor_loadings_df
```

**Insights from Factor Analysis**

In this factor analysis, we've identified five distinct factor groups, each characterized by a unique combination of player attributes. These factor groups provide valuable insights into player performance. Let's assign meaningful names to each factor group:

**Factor 1 - `Technical Prowess`:**

**Attributes:** crossing, finishing, short passing, volleys, dribbling, curve, volleys, free-kick accuracy, long passing, ball control, shot power, long shots, and positioning.

**Description:** Factor 1 represents a player's technical skills and ability to control and manipulate the ball. It encompasses attributes that are essential for attacking and creative roles on the field, such as wingers, forwards, and midfield playmakers. Players with high Factor 1 scores excel in dribbling, passing, and scoring goals.


**Factor 2 - `Defensive Abilities`:**

**Attributes:** standing_tackle, sliding_tackle, interceptions, and aggression

**Description:** Factor 2 represents a player's defensive abilities. It includes attributes related to standing tackle, sliding tackle, interceptions, and aggression. A high factor 2 score indicates strong defensive skills, making the player proficient in preventing the opponent from advancing and regaining possession.

**Factor 3 - `Aerial Dominance`**
This encompasses attributes related to a player's prowess in heading the ball accurately and ball control, which is a crucial skill in football, particularly for players in positions like center-backs, central midfielders, or strikers who are involved in aerial duels, scoring headers, or defending against them.

**Factor 4 - `Speed and Agility`:**

**Attributes:** acceleration, sprint speed, agility, shot power, and reactions.

**Description:** Factor 2 signifies a player's agility and quickness. It includes attributes that are vital for players who need to make rapid movements and quick decisions on the field, such as wingers and agile midfielders. High Factor 2 scores suggest excellent speed and responsiveness.

**Factor 5 - `Physical Dominance`:**

This indicates that it encompasses attributes related to a player's aggression and strength on the field. Players who score high in this factor are likely to be physically dominant, demonstrating their ability to win challenges, hold off opponents, and maintain a strong presence during matches.
